Esta porción de código se encarga de obtener los resultados de los juegos de la Premier League desde el año 1889 al año 2023. Primero, se crea una lista llamada "anios" con todos los años desde 1889 al 2023. Luego, se importa la librería pandas y se crea un dataframe vacío llamado "df_total" para guardar los resultados.

Se itera sobre la lista de años y se crea una lista con todas las jornadas para cada año. Si el año es 2023, las jornadas van del 1 al 26, y si no, van del 1 al 39. Se itera sobre la lista de jornadas y se crea la url para cada jornada. Se hace una petición, se extrae el contenido y se usa BeautifulSoup para obtener los resultados. Se comprueba si la lista está vacía y se eliminan algunos elementos. También se eliminan algunos elementos que contienen información innecesaria. Se agrega la columna de jornada y año al dataframe y se concatenan los resultados. Finalmente, se reordenan las columnas y se guardan los resultados.

```python
# Creando una lista con todos los años
anios = list(range(2000,2024))

# Importando la librería pandas
import pandas as pd

# Creando un dataframe vacío para guardar los resultados
df_total = pd.DataFrame()

# Iterando sobre la lista de años
for anio in anios:
    
    # Creando una lista con todas las jornadas
    if anio == 2023:
        jornadas = list(range(1,26))
    else:
        jornadas = list(range(1,39))
    
    # Iterando sobre la lista de jornadas
    for jornada in jornadas:
    
        # Creando la url para cada jornada
        url = 'https://es.besoccer.com/competicion/resultados/premier/{}/grupoall/jornada{}'.format(anio, jornada)
        req = requests.get(url)
        soup = BeautifulSoup(req.text, 'html.parser')
        divs = soup.find_all("div", {'class': ['panel-body p0 match-list-new']}) 
        texts = [div.text for div in divs]
        data = [text.split("\n") for text in texts]
        data = [[x for x in text if x != ''] for text in data]
        if len(data) == 0: # Agregando la línea para comprobar si la lista está vacía
            continue
        if len(data) > 0:  # Agregando la línea para comprobar si la lista está vacía
            for i, val in enumerate(data[0]):
                if val == "Apl":
                    del data[0][i:i+6]
                if ":" in val:
                    del data[0][i-2:i+3]
        if len(data[0]) > 0: # Agregando la línea para comprobar si la lista está vacía
            for i, val in enumerate(data[0]):
                if val == "Fin":
                    del data[0][i]
        if len(data[0]) > 0: # Agregando la línea para comprobar si la lista está vacía
            for i, val in enumerate(data[0]):
                if "-" in val:
                    result = val.split("-")
                    del data[0][i]
                    data[0].insert(i, result[0])
                    data[0].insert(i+1, result[1])
        if len(data[0]) > 0: # Agregando la línea para comprobar si la lista está vacía
            cols = ["Competencia", "home_team", "score_home", "score_away", "away_team", "date"]
            rows = []

            for i, val in enumerate(data[0]):
                if i % 6 == 0:
                    row = [val]
                else:
                    row.append(val)
                if len(row) == 6:
                    rows.append(row)

            df = pd.DataFrame(rows, columns=cols)
        
            # Agregando las columnas Jornada y Año al dataframe
            df['Jornada'] = jornada
            df['Año'] = anio
        
            # Concatenando los resultados
            df_total = pd.concat([df_total, df])

# Reordenando las columnas
df_total = df_total[['Año','Jornada','Competencia', 'home_team','score_home', 'score_away', 'away_team', 'date']]
```
